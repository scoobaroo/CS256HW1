```python
import pandas as pd
import sympy as sp
import numpy as np
x1 = sp.Symbol('x1')
x2 = sp.Symbol('x2')
y = 1.5*np.power(x1,2) + np.power(x2,2) - 2*x1*x2 + 2*np.power(x1,3) + 0.5*np.power(x1,4)
def f(x1,x2):
    return 1.5*np.power(x1,2) + np.power(x2,2) - 2*x1*x2 + 2*np.power(x1,3) + 0.5*np.power(x1,4)
x1prime = y.diff(x1)
x2prime = y.diff(x2)
print(x1prime)
print(x2prime)
def x1prime(x1,x2):
    return 2.0*x1**3 + 6*x1**2 + 3.0*x1 - 2*x2
def x2prime(x1,x2):
    return -2*x1 + 2*x2
```

```python
def gradient_descent(maxIterations,eta,theta):
    exit = False
    iter = 0
    x1 = 1
    x2 = -3
    while (exit==False):
        print("Iteration", iter, ": x1:", format(x1,'.4f'), ", x2:", format(x2,'.4f'), ", f(x1,x2): ", format(f(x1,x2),'.4f'), ", x1prime:", format(x1prime(x1,x2),'.4f'),", x2prime: ",
              format(x2prime(x1,x2),'.4f'))
        iter+=1
        if(iter>maxIterations):
            print("Max Iterations reached.")
            exit=True
        x1 = x1 - eta*x1prime(x1,x2)
        x2 = x2 - eta*x2prime(x1,x2)
        if (x1prime(x1,x2)**2 + x2prime(x1,x2)**2 )**0.5 < theta:
            print("Theta threshold reached.")
            exit= True
gradient_descent(1000,0.01,0.1)
```

```python
# read data from Assignment1Data.txt
data = pd.read_csv('Assignment1Data.txt', sep=" ", header=None)
data.columns = ["x1", "x2", "y"]

# convert Y to one hot representation
YoneHot = np.zeros((2000,2))
for i in range(2000):
    Yinput = data.loc[[i]][['y']]
    if Yinput['y'].item() == 0:
        YoneHot[i] = [1,0]
    elif Yinput['y'].item() == 1:
        YoneHot[i] = [0,1]
print(YoneHot)
```

```python
trainingSetSize = len(data) # training set size
input_dim = 2 # input layer dimensionality
output_dim = 2 # output layer dimensionality
epsilon = 0.01 # learning rate for gradient descent
# Helper function to evaluate the total loss on the dataset
def calculate_loss(model,X,yonehot):
    W1, W2 = model['W1'], model['W2']
    # Forward propagation
    z1 = X.dot(W1)
    a1 = h(z1)
    z2 = a1.dot(W2)
    softmax_score = softmax(z2)
    return softmax_score
    
# - num_passes: Number of passes through the training data for gradient descent
# - print_loss: If True, print the loss every 1000 iterations
def build_model(theta, epsilon, num_passes=2000):
    # Initialize the parameters to random values. We need to learn these.
    np.random.seed(0)
    W1 = np.random.randn(2, 5) / np.sqrt(2)
    b1 = np.zeros((1, 5))
    W2 = np.random.randn(5, 2) / np.sqrt(5)
    b2 = np.zeros((1, 2))
    exit = False
    model = { 'W1': W1, 'W2': W2 }
    # This is what we return at the end
    # Gradient descent. For each batch...
    for i in range(len(data)):
        # Forward propagation
        X = data.loc[[i]][['x1','x2']]
        Yinput = data.loc[[i]][['y']]
        W1, W2 = model['W1'], model['W2']
        # Forward propagation
        z1 = X.dot(W1)
        a1 = h(z1)
        z2 = a1.dot(W2)
        softmax_score = softmax(z2)
        loss = np.zeros((2,1))
        if YoneHot[i][0] == 1:
            loss[0] = softmax_score - 1
            loss[1] = softmax_score
        else:
            loss[0] = softmax_score
            loss[1] = softmax_score - 1
        if(i > num_passes):
            print("Max Iterations reached.")
            exit = True
            return
        if np.sqrt(loss[0]**1 + loss[1]**2) < theta:
            print("Theta threshold reached.")
            exit = True
            return
        print("Loss at iteration %i " % i , ": ", loss)

        dW2 = loss.T.dot(W2.T)
        print("dW2: ",dW2)
        dW1 = hprime(dW2)
        print("dW1: ", dW1)
        #feed into derivative of sigmoid activator
        # Gradient descent parameter update
        W1 += -epsilon * dW1
        W2 += -epsilon * dW2.T
        # Assign new parameters to the model
        model = { 'W1': W1, 'W2': W2}
        # Optionally print the loss.
        calculate_loss(model,X,YoneHot[i])
    return model

build_model(0.01,0.1,200)
```
