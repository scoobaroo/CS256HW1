```python
import numpy as np
import pandas as pd
import sympy as sp
x1 = sp.Symbol('x1')
x2 = sp.Symbol('x2')
y = 1.5*np.power(x1,2) + np.power(x2,2) - 2*x1*x2 + 2*np.power(x1,3) + 0.5*np.power(x1,4)
def f(x1,x2):
    return 1.5*np.power(x1,2) + np.power(x2,2) - 2*x1*x2 + 2*np.power(x1,3) + 0.5*np.power(x1,4)
x1prime = y.diff(x1)
x2prime = y.diff(x2)
print(x1prime)
print(x2prime)
def x1prime(x1,x2):
    return 2.0*x1**3 + 6*x1**2 + 3.0*x1 - 2*x2
def x2prime(x1,x2):
    return -2*x1 + 2*x2
```

```python
def gradient_descent(maxIterations,eta,theta):
    exit = False
    iter = 0
    x1 = 1
    x2 = -3
    while (exit==False):
        print("Iteration", iter, ": x1:", format(x1,'.2f'), ", x2:", format(x2,'.2f'), ", f(x1,x2): ", f(x1,x2), ", x1prime:", x1prime(x1,x2),", x2prime: ",x2prime(x1,x2))
        
        iter+=1
        if(iter>maxIterations):
            exit=True
        x1 = x1 - eta*x1prime(x1,x2)
        x2 = x2 - eta*x2prime(x1,x2)
        if ((x1prime(x1,x2)**2 + x2prime(x1,x2)**2 )**0.5 < theta ):
            exit= True
gradient_descent(1000,0.00129,0.1)
```
