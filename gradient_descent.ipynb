```python
import pandas as pd
import sympy as sp
import numpy as np
x1 = sp.Symbol('x1')
x2 = sp.Symbol('x2')
y = 1.5*np.power(x1,2) + np.power(x2,2) - 2*x1*x2 + 2*np.power(x1,3) + 0.5*np.power(x1,4)
def f(x1,x2):
    return 1.5*np.power(x1,2) + np.power(x2,2) - 2*x1*x2 + 2*np.power(x1,3) + 0.5*np.power(x1,4)
x1prime = y.diff(x1)
x2prime = y.diff(x2)
print(x1prime)
print(x2prime)
def x1prime(x1,x2):
    return 2.0*x1**3 + 6*x1**2 + 3.0*x1 - 2*x2
def x2prime(x1,x2):
    return -2*x1 + 2*x2
```

```python
def gradient_descent(maxIterations,eta,theta):
    exit = False
    iter = 0
    x1 = 1
    x2 = -3
    while (exit==False):
        print("Iteration", iter, ": x1:", format(x1,'.4f'), ", x2:", format(x2,'.4f'), ", f(x1,x2): ", format(f(x1,x2),'.4f'), ", x1prime:", format(x1prime(x1,x2),'.4f'),", x2prime: ",
              format(x2prime(x1,x2),'.4f'))
        iter+=1
        if(iter>maxIterations):
            print("Max Iterations reached.")
            exit=True
        x1 = x1 - eta*x1prime(x1,x2)
        x2 = x2 - eta*x2prime(x1,x2)
        if (x1prime(x1,x2)**2 + x2prime(x1,x2)**2 )**0.5 < theta:
            print("Theta threshold reached.")
            exit= True
gradient_descent(1000,0.01,0.1)
```

```python
# read data from Assignment1Data.txt
data = pd.read_csv('Assignment1Data.txt', sep=" ", header=None)
data.columns = ["x1", "x2", "y"]

# convert Y to one hot representation
YoneHot = np.zeros((2000,2))
for i in range(2000):
    Yinput = data.loc[[i]][['y']]
    if Yinput['y'].item() == 0:
        YoneHot[i] = [1,0]
    elif Yinput['y'].item() == 1:
        YoneHot[i] = [0,1]
print(YoneHot)
```

```python
def h(x):
    return np.tanh(x)
def hprime(x):
    return 1-(np.tanh(x)**2)
def softmax(input):
    softmax = np.zeros((2,1))
    softmax[0] = np.exp(input[0])/ (np.exp(input[0])+np.exp(input[1]))
    softmax[1] = np.exp(input[1])/ (np.exp(input[0])+np.exp(input[1]))
    return softmax
```

```python
W1 = np.random.randn(2, 5) / np.sqrt(2)
W2 = np.random.randn(5, 2) / np.sqrt(5)
def predict(model, x):
    W1, W2 = model['W1'], model['W2']
    z1 = x.dot(W1)
    a1 = h(z1)
    z2 = a1.dot(W2)
    softmax_score = softmax(z2)
    return softmax_score
```

```python
input_dim = 2 # input layer dimensionality
output_dim = 2 # output layer dimensionality
epsilon = 0.01 # learning rate for gradient descent    
# - num_passes: Number of passes through the training data for gradient descent
# - print_loss: If True, print the loss every 1000 iterations
numberCorrect = 0
def build_model(theta, eta, num_passes=2000, debug=True):
    global numberCorrect
    # Initialize the parameters to random values. We need to learn these.
    np.random.seed(0)
    W1 = np.random.randn(2, 5) / np.sqrt(2)
    W2 = np.random.randn(5, 2) / np.sqrt(5)
    #set exit = False unless stopping criteria is met
    exit = False
    model = { 'W1': W1, 'W2': W2 }
    # This is what we return at the end
    # Gradient descent. For each batch...
    for i in range(1,len(data+1)):
        X = data.loc[[i]][['x1','x2']]
        Y = data.loc[[i]]['y']
        # Forward pass
        input_to_hidden_layer = X.dot(W1) # multiply X with W1
        output_from_hidden_layer = h(input_to_hidden_layer) # feed z1 into sigmoid activation function
        a1prime = hprime(input_to_hidden_layer) # store derivative of sigmoid activation function for use later
        z2 = output_from_hidden_layer.dot(W2) # calculate output of hidden layer by multiplying with W2
        softmax_score = softmax(z2) #softmax normalization
        if debug: print("softmax_score at iteration %i" % i , ": ", softmax_score)
        loss = np.zeros([2])
        prediction = np.round(softmax_score[0])
        if YoneHot[i][0] == 1:
            loss[0] = softmax_score[0] - 1
            loss[1] = softmax_score[0]
        else:
            loss[0] = softmax_score[1]
            loss[1] = softmax_score[1] - 1
        if YoneHot[i][0] == 1 and prediction == 1:
            numberCorrect += 1
        elif YoneHot[i][1] == 1 and prediction == 0:
            numberCorrect += 1
        if debug: print("Loss at iteration %i " % i , ": ", loss)
        if(i > num_passes):
            print("Max Iterations reached.")
            model = { 'W1': W1, 'W2': W2}
            exit = True
            return model
        if np.sqrt(loss[0]**2 + loss[1]**2) < theta:
            print("Theta threshold reached.")
            model = { 'W1': W1, 'W2': W2}
            exit = True
            return model
        # Backward Propagation
        # first step is to multiply loss with W2 to obtain dW2
        loss = loss.T
        dW2 = np.dot(loss,W2.T)
        # take outer product of dW2 and loss to obtain delta W2 for adjusting W2
        deltaW2 = np.outer(dW2,loss)
        # take outer product of a1prime (derivative stored in forward pass) and loss to obtain delta W1 for adjusting W1
        dW1 = np.multiply(dW2,a1prime)
        deltaW1 = np.outer(X,dW1)
        #  weight update for W1 and W2
        W2 += -eta * deltaW2
        W1 += -eta * deltaW1 
        #resetting deltaW1 and deltaW2
        deltaW1 = np.zeros(W1.shape)
        deltaW2 = np.zeros(W2.shape)
        # Assign new parameters to the model
        model = { 'W1': W1, 'W2': W2}
        accuracy = numberCorrect / i
        print("accuracy: ", accuracy)
        print("numberCorrect: ", numberCorrect)
    return model
m = build_model(0.005, 0.02, 2000, True)
m
```
